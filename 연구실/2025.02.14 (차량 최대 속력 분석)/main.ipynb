```python
import matplotlib.pyplot as plt

#  
files = ["resources/avgSpeed-10-high.txt", "resources/avgSpeed-20-high.txt", "resources/avgSpeed-30-high.txt"]
labels = ['10km/h', '20km/h', '30km/h']  #     

#  
plt.figure(figsize=(10, 6))

for i, file in enumerate(files):
    times = []
    speeds = []
    
    with open(file, "r") as f:
        for line in f:
            time, speed = map(float, line.strip().split(","))
            times.append(time)
            speeds.append(speed)
    
    plt.plot(times, speeds, linestyle='-', label=labels[i])  # labels[i] 

#  
plt.xlabel("Time (Seconds)")
plt.ylabel("Speed (km/h)")
plt.title("High Density")
plt.legend()
plt.grid(True)
plt.show()
```
