```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc

#   
if platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')

# 
speed = ['(10km/h)', '(25km/h)', '(50km/h)']
low_density_existing = [59, 63.125, 65.25]
low_density_proposed = [70.5, 68.125, 68.625]
middle_density_existing = [64, 59.375, 50.5]
middle_density_proposed = [75.375, 63.125, 63.125]
high_density_existing = [70.5, 67.25, 59.75]
high_density_proposed = [80, 70.5, 65.875]

#   
plt.figure(figsize=(8, 5))

# 
plt.plot(speed, low_density_existing, 'b--o', label=' -  AODV')
plt.plot(speed, low_density_proposed, 'bo-', label=' - ')

# 
plt.plot(speed, middle_density_existing, 'g--s', label=' -  AODV')
plt.plot(speed, middle_density_proposed, 'gs-', label=' - ')

# 
plt.plot(speed, high_density_existing, 'r--^', label=' -  AODV')
plt.plot(speed, high_density_proposed, 'r-^', label=' - ')

plt.xlabel('')
plt.ylabel('Packet Delivery Ratio (%)')
plt.ylim(50, 100)

#   ( ->  ->  )
plt.legend(loc='best')
plt.grid(True)

#  
plt.tight_layout()
plt.show()
```
