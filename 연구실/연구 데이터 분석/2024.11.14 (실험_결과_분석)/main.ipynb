```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc

# 한글 폰트 설정
if platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')

# 데이터
speed = ['저속(10km/h)', '중간(25km/h)', '고속(50km/h)']
low_density_existing = [59, 63.125, 65.25]
low_density_proposed = [70.5, 68.125, 68.625]
middle_density_existing = [64, 59.375, 50.5]
middle_density_proposed = [75.375, 63.125, 63.125]
high_density_existing = [70.5, 67.25, 59.75]
high_density_proposed = [80, 70.5, 65.875]

# 그래프 크기 설정
plt.figure(figsize=(8, 5))

# 저밀도
plt.plot(speed, low_density_existing, 'b--o', label='저밀도 - 기존 AODV')
plt.plot(speed, low_density_proposed, 'bo-', label='저밀도 - 제안기법')

# 중밀도
plt.plot(speed, middle_density_existing, 'g--s', label='중밀도 - 기존 AODV')
plt.plot(speed, middle_density_proposed, 'gs-', label='중밀도 - 제안기법')

# 고밀도
plt.plot(speed, high_density_existing, 'r--^', label='고밀도 - 기존 AODV')
plt.plot(speed, high_density_proposed, 'r-^', label='고밀도 - 제안기법')

plt.xlabel('속도')
plt.ylabel('Packet Delivery Ratio (%)')
plt.ylim(50, 100)

# 범례 추가 (저밀도 -> 중밀도 -> 고밀도 순서)
plt.legend(loc='best')
plt.grid(True)

# 그래프 출력
plt.tight_layout()
plt.show()
```
