```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc

# 한글 폰트 설정
if platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')

# 데이터
speed = ['저속(10km/h)', '중간(25km/h)', '고속(50km/h)']

# PDR 평균값
low_density_existing = [59, 63.125, 65.25]
low_density_proposed = [70.5, 68.125, 68.625]
middle_density_existing = [64, 59.375, 50.5]
middle_density_proposed = [75.375, 63.125, 63.125]
high_density_existing = [70.5, 67.25, 59.75]
high_density_proposed = [80, 70.5, 65.875]

# Min/Max 값
low_density_min = [43, 15, 15]
low_density_max = [99, 99, 99]
middle_density_min = [34, 18, 16]
middle_density_max = [99, 99, 99]
high_density_min = [27, 17, 11]
high_density_max = [99, 99, 97]

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 평균값 그래프
plt.plot(speed, low_density_existing, 'b--o', label='저밀도 - 기존 AODV')
plt.plot(speed, low_density_proposed, 'bo-', label='저밀도 - 제안기법')
plt.plot(speed, middle_density_existing, 'g--s', label='중밀도 - 기존 AODV')
plt.plot(speed, middle_density_proposed, 'gs-', label='중밀도 - 제안기법')
plt.plot(speed, high_density_existing, 'r--^', label='고밀도 - 기존 AODV')
plt.plot(speed, high_density_proposed, 'r-^', label='고밀도 - 제안기법')

# Min/Max 범위 표시 - 캔들 스타일, 각 밀도별로 약간씩 위치를 이동하여 겹침 방지
offsets = [-0.25, 0, 0.25]  # 저밀도, 중밀도, 고밀도의 위치 조정

for i in range(len(speed)):
    # 저밀도 캔들 스타일
    plt.fill_betweenx(
        [low_density_min[i], low_density_max[i]], i + offsets[0] - 0.1, i + offsets[0] + 0.1,
        color='blue', alpha=0.3, label='' if i > 0 else '저밀도 범위'
    )
    # 중밀도 캔들 스타일
    plt.fill_betweenx(
        [middle_density_min[i], middle_density_max[i]], i + offsets[1] - 0.1, i + offsets[1] + 0.1,
        color='green', alpha=0.3, label='' if i > 0 else '중밀도 범위'
    )
    # 고밀도 캔들 스타일
    plt.fill_betweenx(
        [high_density_min[i], high_density_max[i]], i + offsets[2] - 0.1, i + offsets[2] + 0.1,
        color='red', alpha=0.3, label='' if i > 0 else '고밀도 범위'
    )

plt.xticks(range(len(speed)), speed)
plt.xlabel('속도', labelpad=15)  # x축 레이블 간격 조정
plt.ylabel('Packet Delivery Ratio (%)', labelpad=20)  # y축 레이블 간격 조정
plt.ylim(0, 100)

# 범례를 그래프 외부로 이동
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)
plt.grid(True)

# 그래프 출력
plt.tight_layout()
plt.show()
```
