```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc

#   
if platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')

#  
speeds = ['(10km/h)', '(25km/h)', '(50km/h)']

#    
data_existing = {
    '': [78.6340312, 81.38538293, 84.95310591],
    '': [57.91666667, 61.37478677, 70.76522902],
    '': [39.44211706, 50.99509206, 53.68549471]
}
data_proposed = {
    '': [43.08549118, 47.29821455, 50.29439533],
    '': [49.25983574, 56.81644068, 57.31163577],
    '': [30.34174094, 38.71710901, 45.02017163]
}

# , ,  
colors = ['blue', 'green', 'red']  #  :  ->  -> 
markers = ['^', 's', 'o']  #  :  ->  -> 
linestyles = ['--', '-']  # : , : 

#  
plt.figure(figsize=(10, 6))

#     
for density, color, marker in zip(data_existing.keys(), colors, markers):
    plt.plot(speeds, data_existing[density], label=f'{density} -  AODV', color=color, marker=marker, linestyle=linestyles[0])
    plt.plot(speeds, data_proposed[density], label=f'{density} - ', color=color, marker=marker, linestyle=linestyles[1])

#  
plt.xlabel(' ')
plt.ylabel('RREP  RERR  (%)')
plt.ylim(0, 100)

#  :   AODV   
handles, labels = plt.gca().get_legend_handles_labels()
sorted_handles_labels = []
for density in ['', '', '']:
    for handle, label in zip(handles, labels):
        if density in label:
            sorted_handles_labels.append((handle, label))

#  
plt.legend(
    [h for h, l in sorted_handles_labels], 
    [l for h, l in sorted_handles_labels], 
    title="  ", 
    loc='lower right', 
    frameon=True
)

#  
plt.grid(True)

#  
plt.tight_layout()
plt.show()
```

```python
from skyfield.api import Topos, load

# TLE  
stations_url = 'https://celestrak.com/NORAD/elements/stations.txt'
satellites = load.tle_file(stations_url)
print("TLE   .")

# (ISS) 
satellite = {sat.name: sat for sat in satellites}['ISS (ZARYA)']

#  
ts = load.timescale()
t = ts.now()

#    
geocentric = satellite.at(t)
subpoint = geocentric.subpoint()

latitude = subpoint.latitude.degrees
longitude = subpoint.longitude.degrees
elevation = subpoint.elevation.km

print(f"  :")
print(f": {latitude}")
print(f": {longitude}")
print(f": {elevation} km")
```

```python
import matplotlib.pyplot as plt
import numpy as np

#  (  AODV   )
densities = ['', '', '']
speeds = ['10km/h', '25km/h', '50km/h']
values = {
    '': {' AODV': [56, 75, 70], '': [70, 84, 84]},
    '': {' AODV': [56, 66, 37], '': [84, 75, 60]},
    '': {' AODV': [47, 76, 80], '': [62, 85, 98]}
}

#    
for density in densities:
    plt.figure(figsize=(8, 5))
    
    #  
    aodv = values[density][' AODV']
    improved = values[density]['']
    x = np.arange(len(speeds))
    
    #  
    bar_width = 0.4
    plt.bar(x - bar_width / 2, improved, width=bar_width, label='', color='dodgerblue', alpha=0.85)
    plt.bar(x + bar_width / 2, aodv, width=bar_width, label=' AODV', color='lightgray', alpha=0.85)
    
    # X    
    plt.xticks(x, [f'{speed}' for speed in speeds], fontsize=11, fontweight='medium')
    plt.xlabel(" ", fontsize=12, fontweight='medium')  # X  
    plt.ylabel('Packet Delivery Ratio (%)', fontsize=12)
    plt.ylim(0, 100)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    
    #  
    plt.legend(fontsize=9, loc='upper right', frameon=True, shadow=False, fancybox=False, borderpad=0.5, edgecolor='gray')
    
    #  
    plt.tight_layout()
    plt.show()
```
