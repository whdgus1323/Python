```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = {
    'node': ['0', '4', '6', '9', '10', '11', '17', '22', '26'],
    'legacy': [75, 76, 63, 35, 84, 82, 31, 82, 69],
    'proposed': [75, 97, 63, 53, 90, 77, 30, 82, 71]
}

df = pd.DataFrame(data)

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

bar_width = 0.35
x = range(len(df['node']))

plt.figure(figsize=(10, 5))
plt.bar([i - bar_width/2 for i in x], df['legacy'], width=bar_width, label='Legacy')
plt.bar([i + bar_width/2 for i in x], df['proposed'], width=bar_width, label='Proposed')

plt.xticks(x, df['node'])
plt.xlabel('Node ID')
plt.ylabel('PDR (%)')
plt.title('노드별 PDR 비교 (Legacy vs Proposed)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
