```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

base_path = "C:/Users/Choe JongHyeon/Desktop/Applied Science/시뮬레이션 결과/log/2025_06_17_AppSci_AODV_V07"
density = "D400"
speed = "S30"
nodes = ["N0", "N2", "N3", "N13", "N14", "N16", "N18", "N24", "N29"]
methods = ["L", "P"]

for node in nodes:
    for method in methods:
        node_path = os.path.join(base_path, method, density, speed, node)
        rt_file = os.path.join(node_path, "routing_table_log.csv")
        if not os.path.exists(rt_file):
            continue
        df_rt = pd.read_csv(rt_file)
        destseq_by_time = df_rt.groupby("Time")["DestSeq"].mean()

        rreq_counts = defaultdict(int)
        rreq_file = os.path.join(node_path, "rreqCount.txt")
        if os.path.exists(rreq_file):
            with open(rreq_file, "r") as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) >= 2 and parts[0].replace('.', '', 1).isdigit():
                        t = round(float(parts[0]))
                        rreq_counts[t] += 1

        pdr_by_time = {}
        pdr_file = os.path.join(node_path, "PDR.txt")
        if os.path.exists(pdr_file):
            with open(pdr_file, "r", encoding="utf-8-sig") as f:
                for line in f:
                    raw_line = line.strip()
                    parts = raw_line.split(":")
                    if len(parts) == 2:
                        time_str, pdr_str = parts[0].strip(), parts[1].strip()
                        try:
                            t = int(time_str.lstrip('\ufeff').strip())
                            pdr = float(pdr_str)
                            if not pd.isna(pdr):
                                pdr_by_time[t] = pdr
                        except ValueError as e:
                            print(f"PDR parse error on line: '{raw_line}' -> {e}")
                            continue
            print("PDR parsed data:", pdr_by_time)

        print(f"DestSeq times: {destseq_by_time.index.min()} ~ {destseq_by_time.index.max()}")
        print(f"RREQ times: {min(rreq_counts.keys(), default=None)} ~ {max(rreq_counts.keys(), default=None)}")
        print(f"PDR times: {min(pdr_by_time.keys(), default=None)} ~ {max(pdr_by_time.keys(), default=None)}")

        times = sorted(set(destseq_by_time.index) | set(rreq_counts.keys()) | set(pdr_by_time.keys()))
        destseq_vals = [destseq_by_time.get(t, None) for t in times]
        rreq_vals = [rreq_counts.get(t, 0) for t in times]
        pdr_vals = [pdr_by_time.get(t, 0) for t in times]

        fig, ax1 = plt.subplots(figsize=(12,6))

        destseq_line, = ax1.plot(times, destseq_vals, label=f"{method} DestSeq", color="tab:blue")
        ax1.set_ylabel("DestSeq Mean", color='tab:blue')
        ax1.tick_params(axis='y', labelcolor='tab:blue')

        ax2 = ax1.twinx()
        rreq_line, = ax2.plot(times, rreq_vals, label="RREQ", linestyle="--", color="tab:orange")
        ax2.set_ylabel("RREQ Count", color='tab:orange')
        ax2.tick_params(axis='y', labelcolor='tab:orange')

        ax3 = ax1.twinx()
        ax3.spines.right.set_position(("outward", 60))
        pdr_line, = ax3.plot(times, pdr_vals, label="PDR", linestyle=":", color="tab:green")
        ax3.set_ylabel("PDR (%)", color='tab:green')
        ax3.tick_params(axis='y', labelcolor='tab:green')
        ax3.set_ylim(0, 100)

        ax1.set_xlabel("Time")
        plt.title(f"{node} - {method}")
        fig.tight_layout()
        fig.legend([destseq_line, rreq_line, pdr_line], [f"{method} DestSeq", "RREQ", "PDR"], loc="upper right")
        plt.show()
```
