```python
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
```

```python
files = ["resource/map0.xml", "resource/map1.xml", "resource/map2.xml"]
colors = ["blue", "red", "green"]  # 그래프 색상 지정
labels = ["Map 0", "Map 1", "Map 2"]

plt.figure(figsize=(10, 6))

# 각 파일에 대해 반복
for file, color, label in zip(files, colors, labels):
    tree = ET.parse(file)
    root = tree.getroot()

    time_steps = []
    vehicle_counts = []

    for timestep in root.findall("timestep"):
        time = float(timestep.get("time"))
        vehicle_count = len(timestep.findall("vehicle"))

        time_steps.append(time)
        vehicle_counts.append(vehicle_count)

    # 그래프 그리기
    plt.plot(time_steps, vehicle_counts, marker="o", linestyle="-", color=color, label=label)

# 그래프 설정
plt.xlabel("Time (seconds)", fontsize=1)
plt.ylabel("Number of Vehicles", fontsize=1)
plt.title("Number of Vehicles Over Time", fontsize=14)
plt.legend()
plt.grid()
plt.show()
```
