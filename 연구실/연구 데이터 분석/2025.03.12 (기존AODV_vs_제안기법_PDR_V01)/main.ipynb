```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

csv_filename = "PDR_file/기존_및_제안기법_데이터.csv"
df = pd.read_csv(csv_filename, encoding="utf-8-sig")

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

categories = ['고속 (30 km/h)', '중속 (20 km/h)', '저속 (10 km/h)']
density_levels = df['밀도'].tolist()

existing = df[['고속_기존', '중속_기존', '저속_기존']].to_numpy()

proposed = df[['고속_제안', '중속_제안', '저속_제안']].to_numpy()

fig, ax = plt.subplots(figsize=(10, 6))

for i, level in enumerate(density_levels):
    ax.plot(categories, existing[i], marker='o', linestyle='-', label=f'기존 - {level}', alpha=0.7)

for i, level in enumerate(density_levels):
    ax.plot(categories, proposed[i], marker='s', linestyle='--', label=f'제안기법 - {level}', alpha=0.7)

ax.set_xlabel('속도 (km/h)')
ax.set_ylabel('Packet Delivery Ratio (PDR) %')
ax.set_title('기존 vs 제안 기법 성능 비교')
ax.set_ylim(70, 100)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

csv_filename = "PDR_file/Existing_vs_Proposed_Data.csv"
df = pd.read_csv(csv_filename, encoding="utf-8-sig")

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

categories = ['High Speed (30 km/h)', 'Medium Speed (20 km/h)', 'Low Speed (10 km/h)']
density_levels = df['Density'].tolist()

existing = df[['High_Existing', 'Medium_Existing', 'Low_Existing']].to_numpy()
proposed = df[['High_Proposed', 'Medium_Proposed', 'Low_Proposed']].to_numpy()

fig, ax = plt.subplots(figsize=(10, 6))

colors = ['blue', 'green', 'red']

for i, level in enumerate(density_levels):
    ax.plot(categories, existing[i], marker='o', linestyle='-', color=colors[i], 
            label=f'Legacy - {level}', alpha=0.7)

for i, level in enumerate(density_levels):
    ax.plot(categories, proposed[i], marker='s', linestyle='--', color=colors[i], 
            label=f'Proposed - {level}', alpha=0.7)

ax.set_xlabel('Speed (km/h)')
ax.set_ylabel('Packet Delivery Ratio (PDR) %')
ax.set_title('Performance Comparison: Legacy AODV vs. Proposed')
ax.set_ylim(70, 100)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
