```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'result.csv'
data = pd.read_csv(file_path)

# Clean the data by removing unnecessary columns
cleaned_data = data.loc[:, ~data.columns.str.contains('^Unnamed')]

# Prepare data for plotting
x = range(len(cleaned_data))  # Assuming the index represents data points
original_pdr = cleaned_data['a']
improved_pdr = cleaned_data['b']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, original_pdr, label='Original PDR', marker='o')
plt.plot(x, improved_pdr, label='Improved PDR', marker='x')

# Add labels and title
plt.xlabel('Data Points')
plt.ylabel('PDR (%)')
plt.title('Comparison of Original and Improved PDR')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'result.csv'
data = pd.read_csv(file_path)

cleaned_data = data.loc[:, ~data.columns.str.contains('^Unnamed')]

pairs = [(cleaned_data[f'a.{i}'] if f'a.{i}' in cleaned_data.columns else cleaned_data['a'],
          cleaned_data[f'b.{i}'] if f'b.{i}' in cleaned_data.columns else cleaned_data['b']) 
         for i in range(7)]

titles = ["Node [0]", "Node [34]", "Node [39]", "Node [13]", "Node [30]", "Node [40]", "Node [18]"]

fig, axes = plt.subplots(4, 2, figsize=(15, 20))

axes = axes.flatten()

for idx, (original_pdr, improved_pdr) in enumerate(pairs):
    x = range(len(original_pdr))
    axes[idx].plot(x, original_pdr, linestyle='-', label=f'Original AODV')
    axes[idx].plot(x, improved_pdr, linestyle='-', label=f'Improved AODV')
    axes[idx].set_ylabel('Packet Delivery Rate (%)')
    axes[idx].set_xlabel('Seconds')
    axes[idx].set_title(titles[idx])
    axes[idx].legend(loc='lower right')
    axes[idx].grid(True)

for j in range(len(pairs), len(axes)):
    fig.delaxes(axes[j])

plt.xlabel('Data Points')
plt.tight_layout()
plt.show()

```
