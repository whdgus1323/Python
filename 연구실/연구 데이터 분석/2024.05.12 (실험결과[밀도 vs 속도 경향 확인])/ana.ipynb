```python
import pandas as pd
import matplotlib.pyplot as plt

file_paths = [
    ' CSV/ .CSV', 
    ' CSV/ .CSV',  
    ' CSV/ .CSV',  
    ' CSV/ .CSV',   
    ' CSV/ .CSV',  
    ' CSV/ .CSV'  
]

dataframes = [pd.read_csv(file_path, encoding='cp949') for file_path in file_paths]

low_density_improved_pdr = [df['P'].mean() for df in dataframes[:3]]
low_density_original_pdr = [df['B'].mean() for df in dataframes[:3]] 
medium_density_improved_pdr = [df['P'].mean() for df in dataframes[3:]] 
medium_density_original_pdr = [df['B'].mean() for df in dataframes[3:]] 

x_labels = ['Low Speed', 'Medium Speed', 'High Speed']

plt.figure(figsize=(8, 5))

plt.plot(x_labels, low_density_improved_pdr, marker='o', label='Proposed AODV - Low Density', linestyle='-')
plt.plot(x_labels, low_density_original_pdr, marker='x', label='AODV - Low Density', linestyle='--')

plt.plot(x_labels, medium_density_improved_pdr, marker='o', label='Proposed AODV - Medium Density', linestyle='-')
plt.plot(x_labels, medium_density_original_pdr, marker='x', label='AODV - Medium Density', linestyle='--')

plt.xlabel('Speed Category')
plt.ylabel('Packet Delivery Ratio (PDR) (%)')
plt.title('PDR by Speed and Density')
plt.legend(loc='lower right')
plt.grid(True)

plt.ylim(0, 100) 
plt.tight_layout()
plt.show()
```
