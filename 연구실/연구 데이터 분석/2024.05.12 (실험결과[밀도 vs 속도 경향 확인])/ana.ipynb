```python
import pandas as pd
import matplotlib.pyplot as plt

file_paths = [
    '실험 CSV/저밀도 저속.CSV', 
    '실험 CSV/저밀도 중간.CSV',  
    '실험 CSV/저밀도 고속.CSV',  
    '실험 CSV/중밀도 저속.CSV',   
    '실험 CSV/중밀도 중간.CSV',  
    '실험 CSV/중밀도 고속.CSV'  
]

dataframes = [pd.read_csv(file_path, encoding='cp949') for file_path in file_paths]

low_density_improved_pdr = [df['P'].mean() for df in dataframes[:3]]
low_density_original_pdr = [df['B'].mean() for df in dataframes[:3]] 
medium_density_improved_pdr = [df['P'].mean() for df in dataframes[3:]] 
medium_density_original_pdr = [df['B'].mean() for df in dataframes[3:]] 

x_labels = ['Low Speed', 'Medium Speed', 'High Speed']

plt.figure(figsize=(8, 5))

plt.plot(x_labels, low_density_improved_pdr, marker='o', label='Proposed AODV - Low Density', linestyle='-')
plt.plot(x_labels, low_density_original_pdr, marker='x', label='AODV - Low Density', linestyle='--')

plt.plot(x_labels, medium_density_improved_pdr, marker='o', label='Proposed AODV - Medium Density', linestyle='-')
plt.plot(x_labels, medium_density_original_pdr, marker='x', label='AODV - Medium Density', linestyle='--')

plt.xlabel('Speed Category')
plt.ylabel('Packet Delivery Ratio (PDR) (%)')
plt.title('PDR by Speed and Density')
plt.legend(loc='lower right')
plt.grid(True)

plt.ylim(0, 100) 
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

data = {
    '속도': ['저속(5km/h)', '중간(10km/h)', '고속(30km/h)'] * 3,
    '밀집도': ['Low Density', 'Low Density', 'Low Density', 'Middle Density', 'Middle Density', 'Middle Density', 'High Density', 'High Density', 'High Density'],
    '제안기법': [    87.57971, 84.15941, 77.94202, 
                    68.79712, 84.15941, 68.64733, 
                    62.66667, 68.22222, 78.88889],
    '기존 AODV': [  77.94202, 77.5138, 78.33332, 
                    64.13043, 73.94202, 68.80836, 
                    67.55556, 71.77778, 77.77778]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(6, 4))

colors = {
    'Low Density': 'blue',
    'Middle Density': 'green',
    'High Density': 'red'
}

for 밀집도 in df['밀집도'].unique():
    subset = df[df['밀집도'] == 밀집도]
    color = colors[밀집도]
    ax.plot(subset['속도'], subset['제안기법'], marker='o', color=color, label=f'{밀집도} - 제안기법')
    ax.plot(subset['속도'], subset['기존 AODV'], marker='s', linestyle='--', color=color, label=f'{밀집도} - 기존 AODV')

ax.set_ylim(45, 100)

ax.set_xlabel('Speed')
ax.set_ylabel('Packet Delivery Rate(%)')
ax.set_title('밀집도별 속도에 따른 PDR 비교')

ax.legend(loc='lower right', bbox_to_anchor=(1, 0), frameon=True, fontsize='small', fancybox=True, shadow=True)


ax.grid(True)

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 설정 (개선과 기존 성능)
data = {
    '버전': ['개선', '기존'],
    'PDR': [84, 77]
}

df = pd.DataFrame(data)

# Bar 그래프 그리기
fig, ax = plt.subplots(figsize=(6, 4))

# 막대 그래프 생성
ax.bar(df['버전'], df['PDR'], color=['blue', 'orange'])

ax.set_xlabel('버전')
ax.set_ylabel('PDR')
ax.set_title('개선 vs 기존 AODV PDR 비교')

# Y축 범위를 100으로 설정
ax.set_ylim(0, 100)

plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (예: 맑은 고딕)
plt.rc('font', family='Malgun Gothic')

# 데이터 설정 (개선과 기존 성능)
data = [
    {'개선': 91, '기존': 70},
    {'개선': 94, '기존': 58},
    {'개선': 99, '기존': 82},
    
    {'개선': 94, '기존': 87},
    {'개선': 80, '기존': 71},
    {'개선': 75, '기존': 15},
    
    {'개선': 99, '기존': 71},
    {'개선': 98, '기존': 58},
    {'개선': 75, '기존': 60},
    {'개선': 98, '기존': 85}
]

# 각각의 데이터를 개별 그래프로 표시
for idx, entry in enumerate(data):
    fig, ax = plt.subplots(figsize=(6, 4))
    
    # 개선과 기존 데이터를 각각 묶어서 바 그래프 생성
    ax.bar(['개선', '기존 AODV'], [entry['개선'], entry['기존']], color=['blue', 'orange'])
    
    ax.set_ylabel('Packet Delivery Rate(%)')
    ax.set_title(f"개선 vs 기존 AODV PDR 비교 (Case {idx+1})")
    
    # Y축 범위를 100으로 설정
    ax.set_ylim(0, 100)
    
    plt.tight_layout()
    plt.show()
```
