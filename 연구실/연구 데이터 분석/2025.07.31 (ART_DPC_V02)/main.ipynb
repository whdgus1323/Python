```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# 정확한 경로
file_path = r"C:\Users\Choe JongHyeon\Desktop\Applied Science\시뮬레이션 결과\Log\2025_07_14_AppSci_AODV_V10\L\D400\S30\N\2-p\routing_table_log.csv"
pdr_path = r"C:\Users\Choe JongHyeon\Desktop\Applied Science\시뮬레이션 결과\Log\2025_07_14_AppSci_AODV_V10\L\D400\S30\N\2-p\PDR.txt"

# 라우팅 로그 불러오기
df = pd.read_csv(file_path)
df = df[df['Lifetime'] > 0]
df['Lifetime'] = df['Lifetime'].astype(float)
df['Density'] = df['Density'].astype(float)

density_lifetime = df.groupby('Density')['Lifetime'].mean().reset_index()

# PDR.txt 처리: 숫자 줄만 필터링
with open(pdr_path, 'r') as f:
    lines = f.readlines()

clean_lines = [
    line for line in lines
    if all(char.isdigit() or char.isspace() or char == '.' for char in line.strip())
]

pdr_df = pd.read_csv(StringIO("".join(clean_lines)), sep=r"\s+|\t+|,", engine='python', names=["Density", "PDR"])
pdr_df['Density'] = pdr_df['Density'].astype(float)
pdr_df['PDR'] = pdr_df['PDR'].astype(float)

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_xlabel("Density")
ax1.set_ylabel("Average Lifetime", color='tab:blue')
ax1.plot(density_lifetime['Density'], density_lifetime['Lifetime'], marker='o', color='tab:blue', label="Lifetime")
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True)

ax2 = ax1.twinx()
ax2.set_ylabel("PDR", color='tab:red')
ax2.plot(pdr_df['Density'], pdr_df['PDR'], marker='s', color='tab:red', label="PDR")
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title("Lifetime and PDR vs Density")
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 파일 경로 지정
file_path = r"C:\Users\Choe JongHyeon\Desktop\Applied Science\시뮬레이션 결과\Log\2025_07_14_AppSci_AODV_V10\L\D400\S30\N\2-p\routing_table_log.csv"
pdr_path = r"C:\Users\Choe JongHyeon\Desktop\Applied Science\시뮬레이션 결과\Log\2025_07_14_AppSci_AODV_V10\L\D400\S30\N\2-p\PDR.txt"

# 라우팅 테이블 로그 불러오기
df = pd.read_csv(file_path)
df = df[df['Lifetime'] > 0]

df['Lifetime'] = df['Lifetime'].astype(float)
df['Density'] = df['Density'].astype(float)
df['AvgRelSpeed'] = df['AvgRelSpeed'].astype(float)

# 밀도별 평균 Lifetime 계산
density_lifetime = df.groupby('Density')['Lifetime'].mean().reset_index()

# PDR.txt 불러오기
# 예: "0 0.85" 형식 또는 "Density,PDR"
try:
    pdr_df = pd.read_csv(pdr_path, sep=r"\s+|,|\t+", engine='python', names=["Density", "PDR"])
    pdr_df = pdr_df.dropna()
    pdr_df['Density'] = pd.to_numeric(pdr_df['Density'], errors='coerce')
    pdr_df['PDR'] = pd.to_numeric(pdr_df['PDR'], errors='coerce')
    pdr_df = pdr_df.dropna()
except Exception as e:
    print("PDR 파일 오류:", e)
    pdr_df = pd.DataFrame(columns=["Density", "PDR"])

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 6))

# Lifetime 축
ax1.set_xlabel("Density")
ax1.set_ylabel("Average Lifetime", color='tab:blue')
ax1.plot(density_lifetime['Density'], density_lifetime['Lifetime'], color='tab:blue', marker='o', label='Lifetime')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True)

# PDR 축
ax2 = ax1.twinx()
ax2.set_ylabel("PDR", color='tab:red')
if not pdr_df.empty:
    ax2.plot(pdr_df['Density'], pdr_df['PDR'], color='tab:red', marker='s', label='PDR', linewidth=2.5)
ax2.tick_params(axis='y', labelcolor='tab:red')

# 범례 표시
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.title("Lifetime and PDR vs Density")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

n0_path = r"C:\Users\Choe JongHyeon\Desktop\Applied Science\시뮬레이션 결과\Log\2025_07_14_AppSci_AODV_V10\L\D400\S30\N\0-p\routing_table_log.csv"
n2_path = r"C:\Users\Choe JongHyeon\Desktop\Applied Science\시뮬레이션 결과\Log\2025_07_14_AppSci_AODV_V10\L\D400\S30\N\2-p\routing_table_log.csv"

def load_log(path):
    df = pd.read_csv(path, sep="\t" if '\t' in open(path).readline() else ",")
    df['Time'] = df['Time'].astype(float).astype(int)
    df['Destination'] = df['Destination'].astype(str)
    df['NextHop'] = df['NextHop'].astype(str)
    df['Lifetime'] = df['Lifetime'].astype(float)
    return df

def compute_lifetime_variance(df):
    avg_lifetime = df.groupby('Time')['Lifetime'].mean()
    delta = avg_lifetime.diff().dropna()
    return np.var(delta)

def compute_update_count(df):
    df_sorted = df.sort_values(['Destination', 'Time'])
    df_shifted = df_sorted.groupby('Destination')[['NextHop', 'Lifetime']].shift(1)
    changed = (df_sorted['NextHop'] != df_shifted['NextHop']) | (df_sorted['Lifetime'] != df_shifted['Lifetime'])
    return changed.sum()

df0 = load_log(n0_path)
df2 = load_log(n2_path)

var_life_0 = compute_lifetime_variance(df0)
var_life_2 = compute_lifetime_variance(df2)
updates_0 = compute_update_count(df0)
updates_2 = compute_update_count(df2)

print("Node 0 - Var(ΔLifetime):", round(var_life_0, 5), "| Update Count:", updates_0)
print("Node 2 - Var(ΔLifetime):", round(var_life_2, 5), "| Update Count:", updates_2)
```
