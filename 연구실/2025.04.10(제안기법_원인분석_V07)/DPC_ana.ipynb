```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

with open('dpc.txt', 'r') as f:
    data = [float(line.strip()) for line in f if line.strip()]

data = np.array(data)

sorted_data = np.sort(data)
cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)

plt.figure(figsize=(8, 5))
plt.plot(sorted_data, cdf, linestyle='-', label='CDF')
plt.xlabel('DPC Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of DPC')
plt.grid(True)

q1, q2, q3 = np.percentile(data, [25, 50, 75])
for q, label in zip([q1, q2, q3], ['Q1', 'Median', 'Q3']):
    plt.axvline(q, color='r', linestyle='--', label=label)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(5, 6))
sns.boxplot(y=data)
plt.title('Boxplot of DPC')
plt.ylabel('DPC Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data, kde=True, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of DPC')
plt.xlabel('DPC Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 경로 설정
pwdP = "/기존"
pwdD = "/고밀도"
pwdS = "/고속"
base_path = "C:/Users/Choe JongHyeon/Desktop/Applied Science/시뮬레이션 결과/Result_V02" + pwdP + pwdS + pwdD

# 노드 목록
node_list = [0, 4, 6, 7, 9, 10, 11, 17, 19, 20]

# 결과 저장용 dict
summary = {}
all_node_speed_cdf = {}
all_node_vehicle_cdf = {}

# 마지막 값만 읽는 함수
def get_last_float(filepath):
    if not os.path.exists(filepath):
        return np.nan
    with open(filepath, 'r') as f:
        lines = f.readlines()
        if not lines:
            return np.nan
        try:
            return float(lines[-1].split(":")[-1].strip())
        except:
            return np.nan

# 파일 읽기 및 데이터 저장
for node_id in node_list:
    node_path = os.path.join(base_path, f"{node_id}")
    vehicle_file = os.path.join(node_path, "vehicle_status.csv")

    avg_speed = avg_density = np.nan
    pdr = np.nan
    rreq_count = 0
    linkbreak_count = 0

    if os.path.exists(vehicle_file):
        df = pd.read_csv(vehicle_file, header=None)
        if df.shape[1] >= 5:
            df.columns = ['time', 'node', 'veh_count', 'speed', 'density']
            avg_speed = df['speed'].mean()
            avg_density = df['density'].mean()

            grouped = df.groupby('node').agg({
                'speed': 'mean',
                'veh_count': 'mean'
            }).reset_index()
            all_node_speed_cdf[node_id] = grouped['speed'].tolist()
            all_node_vehicle_cdf[node_id] = grouped['veh_count'].tolist()

    pdr = get_last_float(os.path.join(node_path, "PDR.txt"))

    rreq_file = os.path.join(node_path, "rreqCount.txt")
    rreq_count = sum(1 for _ in open(rreq_file)) if os.path.exists(rreq_file) else 0

    lb_file = os.path.join(node_path, "LinkBreakCount.txt")
    linkbreak_count = sum(1 for _ in open(lb_file)) if os.path.exists(lb_file) else 0

    summary[node_id] = {
        'Node': node_id,
        'PDR': pdr,
        'RREQ_Count': rreq_count,
        'LinkBreak_Count': linkbreak_count,
        'AvgSpeed': avg_speed,
        'AvgDensity': avg_density
    }

# 요약 결과 저장
df_summary = pd.DataFrame(summary.values())
df_summary.to_csv("summary_result.csv", index=False)

# 각 노드별 CDF + KDE 시각화 함수
def plot_cdf_with_kde(data_list, node_id, label):
    data = np.array(data_list)
    data = data[~np.isnan(data)]
    if len(data) == 0:
        print(f"[경고] Node[{node_id}] {label} 시각화할 데이터 없음")
        return

    q25 = np.percentile(data, 25)
    q50 = np.percentile(data, 50)
    q75 = np.percentile(data, 75)
    q90 = np.percentile(data, 90)

    sorted_data = np.sort(data)
    yvals = np.arange(1, len(sorted_data)+1) / float(len(sorted_data))

    plt.figure(figsize=(10, 4))

    # CDF
    plt.subplot(1, 2, 1)
    plt.plot(sorted_data, yvals, label='CDF', color='blue')
    for q, label_txt, color in zip([q25, q50, q75, q90], ['25%', 'Median', '75%', '90%'],
                                   ['green', 'red', 'orange', 'purple']):
        plt.axvline(x=q, linestyle='--', color=color, label=f'{label_txt}: {q:.2f}')
    plt.xlabel(label)
    plt.ylabel('CDF')
    plt.title(f"Node[{node_id}] {label} CDF")
    plt.legend()
    plt.grid(True)

    # KDE
    plt.subplot(1, 2, 2)
    sns.kdeplot(data, fill=True, color='skyblue')
    plt.xlabel(label)
    plt.ylabel('Density')
    plt.title(f"Node[{node_id}] {label} KDE")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# 전체 KDE 시각화 함수
def plot_kde_all_nodes(data_dict, title, xlabel):
    plt.figure(figsize=(10, 5))
    for node_id, values in data_dict.items():
        values = np.array(values)
        values = values[~np.isnan(values)]
        if len(values) > 0:
            sns.kdeplot(values, label=f"Node[{node_id}]")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 각 노드별 시각화 실행
for node_id in node_list:
    if node_id in all_node_speed_cdf:
        plot_cdf_with_kde(all_node_speed_cdf[node_id], node_id, "Average Speed (km/h)")
    if node_id in all_node_vehicle_cdf:
        plot_cdf_with_kde(all_node_vehicle_cdf[node_id], node_id, "Average Vehicle Count")

# 전체 노드 KDE 비교
plot_kde_all_nodes(all_node_speed_cdf, "노드별 평균 속력 KDE", "Average Speed (km/h)")
plot_kde_all_nodes(all_node_vehicle_cdf, "노드별 평균 차량 수 KDE", "Average Vehicle Count")
```
