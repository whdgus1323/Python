```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc

if platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')

speed = ['(10km/h)', '(25km/h)', '(50km/h)']
low_density_existing = [59, 63.125, 65.25]
low_density_proposed = [70.5, 68.125, 68.625]
middle_density_existing = [64, 59.375, 50.5]
middle_density_proposed = [75.375, 63.125, 63.125]
high_density_existing = [70.5, 67.25, 59.75]
high_density_proposed = [80, 70.5, 65.875]

plt.figure(figsize=(8, 5))

plt.plot(speed, low_density_proposed, 'bo-', label='Low Density - ')
plt.plot(speed, low_density_existing, 'b--o', label='Low Density -  AODV')

plt.plot(speed, middle_density_proposed, 'go-', label='Middle Density - ')
plt.plot(speed, middle_density_existing, 'g--o', label='Middle Density -  AODV')

plt.plot(speed, high_density_proposed, 'ro-', label='High Density - ')
plt.plot(speed, high_density_existing, 'r--o', label='High Density -  AODV')

plt.title('   PDR ')
plt.xlabel('Speed')
plt.ylabel('Packet Delivery Rate(%)')
plt.ylim(50, 100)

plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Speed categories and density data
speeds = ['(10km/h)', '(25km/h)', '(50km/h)']

# Standard deviation data for each density and protocol (Proposed and AODV)
data_aodv_std = {
    'Low Density': [16.88617355, 24.61960601, 24.83517091],
    'Middle Density': [18.83765227, 28.66026368, 29.68886275],
    'High Density': [27.06077815, 27.45775977, 32.29661636]
}
data_proposed_std = {
    'Low Density': [12.29401713, 22.05472091, 21.3938409],
    'Middle Density': [13.28734629, 25.29504469, 25.62051354],
    'High Density': [16.3270153, 26.55452611, 28.50281941]
}

# Colors and markers for each line
colors = ['blue', 'green', 'red']
markers = ['o', 's', '^']
linestyles = ['--', '-']

# Plotting
plt.figure(figsize=(10, 6))
for i, (density, color, marker) in enumerate(zip(['Low Density', 'Middle Density', 'High Density'], colors, markers)):
    plt.plot(speeds, data_proposed_std[density], label=f'{density} - ', color=color, marker=marker, linestyle=linestyles[0])
    plt.plot(speeds, data_aodv_std[density], label=f'{density} -  AODV', color=color, marker=marker, linestyle=linestyles[1])

# Labels and title
plt.xlabel('Speed')
plt.ylabel('Standard Deviation (%)')
plt.title('    ')
plt.ylim(10, 40)
plt.legend(loc='best')

# Show grid for better readability
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([
    [31.125, 23, 29.875],
    [11.75, 23.625, 22.125], 
    [5.125, 31.125, 33.5] 
])

speed_labels = ['(10km/h)', '(25km/h)', '(50km/h)']
density_labels = ['', '', '']
colors = ['blue', 'green', 'red'] 

plt.figure(figsize=(10, 6))

for i in range(len(density_labels)):
    plt.plot(speed_labels, data[i], marker='o', linestyle='-', color=colors[i], label=density_labels[i])

plt.xlabel('Speed')
plt.ylabel('  ')
plt.title('      ')
plt.ylim(0, 40)
plt.legend(loc='best', title='')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
densities = ['', '', '']
categories = ['', '', '']
values = {
    '': {'': [56, 75, 70], ' AODV': [70, 84, 84]},
    '': {'': [56, 66, 37], ' AODV': [84, 75, 60]},
    '': {'': [47, 76, 80], ' AODV': [62, 85, 98]}
}

# Plotting each density and speed separately with ' AODV' first and '' second
for density in densities:
    for i, category in enumerate(categories):
        plt.figure()
        # Plot ' AODV' first, then ''
        plt.bar([' AODV', ''], [values[density][' AODV'][i], values[density][''][i]], color=['orange', 'blue'])
        plt.ylim(0, 100)  # Set the y-axis limit to 100 for consistency
        plt.ylabel('Packet Delivery Rate(%)')
        plt.title(f'{density} - {category}  AODV vs  PDR ')
        plt.show()
```
