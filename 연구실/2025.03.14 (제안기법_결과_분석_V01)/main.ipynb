```python
import pandas as pd
import os
import re

density_list = ['고밀도', '중밀도', '저밀도']
speed_list = ['고속', '중속', '저속']
mode_list = ['기존', '제안기법']

node_dict = {
    '고밀도': {'고속': [6, 7, 18, 21, 29, 33, 36, 44, 46],
              '중속': [6, 7, 18, 16, 33, 36, 29, 37, 38],
              '저속': [6, 7, 18, 33, 36, 29, 37, 46, 22]},

    '중밀도': {'고속': [6, 7, 17, 20, 24, 26, 27, 29, 36, 37],
              '중속': [6, 7, 17, 20, 24, 26, 27, 29, 36, 37],
              '저속': [6, 17, 18, 24, 26, 27, 29, 36, 37, 42]},

    '저밀도': {'고속': [6, 7, 17, 18, 20, 21, 22, 24, 25, 27],
              '중속': [6, 7, 17, 18, 20, 21, 22, 24, 29, 33],
              '저속': [7, 17, 18, 20, 21, 22, 24, 25, 27, 31]}
}

data = []
additional_data = []

additional_files = [
    "PDR.txt", "rerr.txt", "rrepCnt.txt", "rrepCount.txt", "rreqCount.txt", "sendPing.txt",
    "success_app.txt", "transmitCount.txt", "collision_log.txt", "createRouteCnt.txt", "LinkBreakCount.txt", "failedCount.txt"
]

for density in density_list:
    for speed in speed_list:
        for mode in mode_list:
            base_path = f"C:/Users/Choe JongHyeon/Desktop/Applied Science/시뮬레이션 결과/Lifetime 결과/{density}/{speed}/{mode}/"

            nodes = node_dict[density][speed]

            for node_id in nodes:
                node_path = os.path.join(base_path, str(node_id))

                if not os.path.exists(node_path):
                    print(f"노드 폴더 없음: {os.path.abspath(node_path)}")
                    continue
                
                file_path = os.path.join(node_path, "aodv_routing_log.csv")
                if not os.path.exists(file_path):
                    print(f"파일 없음: {os.path.abspath(file_path)}")
                else:
                    try:
                        df = pd.read_csv(file_path, nrows=500)
                    except Exception as e:
                        print(f"파일 로딩 실패: {os.path.abspath(file_path)}, 에러: {e}")
                        continue

                    def extract_node(x):
                        match = re.search(r'\[(\d+)\]', str(x))
                        return int(match.group(1)) if match else -1

                    if 'Node' in df.columns:
                        df['Node'] = df['Node'].apply(extract_node)

                    if 'New_LifeTime' in df.columns and 'Previous_LifeTime' in df.columns:
                        df['LifeTime_Change'] = df['New_LifeTime'] - df['Previous_LifeTime']
                    else:
                        print(f"LifeTime 컬럼 없음: {os.path.abspath(file_path)}")
                        continue

                    for row in df.itertuples(index=False):
                        data.append([density, speed, mode, node_id, row.New_LifeTime, row.LifeTime_Change])

                for file_name in additional_files:
                    additional_file_path = os.path.join(node_path, file_name)

                    if not os.path.exists(additional_file_path):
                        print(f"파일 없음: {os.path.abspath(additional_file_path)}")
                        continue

                    try:
                        with open(additional_file_path, 'r', encoding='utf-8', errors='replace') as f:
                            lines = f.readlines()

                            if file_name.endswith(".csv"):  
                                df_add = pd.read_csv(additional_file_path)
                                for row in df_add.itertuples(index=False):
                                    additional_data.append([density, speed, mode, node_id, file_name] + list(row))

                            elif file_name in ["success_app.txt", "transmitCount.txt", "collision_log.txt"]:  
                                df_add = pd.read_csv(additional_file_path, sep="\s+", header=None)
                                for row in df_add.itertuples(index=False):
                                    additional_data.append([density, speed, mode, node_id, file_name] + list(row))
                                #print(df_add)

                            else:  
                                for line in lines:
                                    additional_data.append([density, speed, mode, node_id, file_name, line.strip()])

                    except Exception as e:
                        print(f"파일 로딩 실패: {os.path.abspath(additional_file_path)}, 에러: {e}")

df_result = pd.DataFrame(data, columns=['Density', 'Speed', 'Method', 'Node', 'LifeTime', 'LifeTime_Change'])
df_additional = pd.DataFrame(additional_data)

print("데이터 로딩 완료")
```

```python
import pandas as pd
import numpy as np

# df_result 요약
if not df_result.empty:
    print("\n===== df_result 요약 =====")
    print(f"총 데이터 개수: {len(df_result)}")
    
    # 밀도별 평균 LifeTime
    density_summary = df_result.groupby("Density")["LifeTime"].agg(["count", "mean", "std"])
    print("\n[밀도별 평균 LifeTime]")
    print(density_summary)

    # 속도별 평균 LifeTime
    speed_summary = df_result.groupby("Speed")["LifeTime"].agg(["count", "mean", "std"])
    print("\n[속도별 평균 LifeTime]")
    print(speed_summary)

    # 기법별 평균 LifeTime
    method_summary = df_result.groupby("Method")["LifeTime"].agg(["count", "mean", "std"])
    print("\n[기법별 평균 LifeTime]")
    print(method_summary)

    # 노드별 평균 LifeTime 및 LifeTime 변화량
    node_summary = df_result.groupby("Node").agg(
        LifeTime_Mean=("LifeTime", "mean"),
        LifeTime_Change_Mean=("LifeTime_Change", "mean"),
        Count=("Node", "count")
    )
    print("\n[노드별 평균 LifeTime 및 LifeTime 변화량]")
    print(node_summary.head(10))  # 상위 10개만 출력

else:
    print("\n===== df_result 요약 =====")
    print("데이터 없음.")

# df_additional 요약
if not df_additional.empty:
    print("\n===== df_additional 요약 =====")
    print(f"총 추가 데이터 개수: {len(df_additional)}")

    # 파일별 데이터 개수
    file_summary = df_additional.groupby("File")["File"].count().sort_values(ascending=False)
    print("\n[파일별 데이터 개수]")
    print(file_summary)

    # 노드별 수집된 파일 개수
    node_file_summary = df_additional.groupby("Node")["File"].count().sort_values(ascending=False)
    print("\n[노드별 수집된 파일 개수]")
    print(node_file_summary.head(10))  # 상위 10개만 출력

else:
    print("\n===== df_additional 요약 =====")
    print("추가 데이터 없음.")
```
